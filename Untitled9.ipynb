```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import  seaborn as sns
%matplotlib inline
df=pd.read_csv('matches.csv')
df.head(10)
```

```python
print('number of  matches played so far:-',df.shape[0])
print('number of seasons:-' ,len(df.season.unique()))
'''now the first question we are going to solve here is \n what is the number of
matches each season'''
```

```python
sns.countplot(x='season',data=df)
plt.show()
```

```python
print('number of matches in each vanue')
plt.figure(figsize=(12,6))
sns.countplot(x='venue',data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
print('number of matches played bby each team')


temp_df = pd.melt(df, id_vars=['id','season'], value_vars=['team1', 'team2'])

plt.figure(figsize=(12,6))
sns.countplot(x='value', data=temp_df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
print('number of wins per team')
sns.countplot(x='winner', data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
print('Champions each season:')
arg_df = df.drop_duplicates(subset=['season'], keep='last')[['season', 'winner']].reset_index(drop=True)
arg_df
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='toss_decision', data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
df["field_win"] = "win"
df["field_win"].ix[df['win_by_wickets']==0] = "loss"
plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='field_win', data=df)
plt.xticks(rotation='vertical')
plt.show()
```
